```python
# Select the first 11 columns
selected_columns = df.iloc[:, 0:11]

# Convert "Yes" to 1 and other values to 0
MD_x = (selected_columns == "Yes").astype(int)

# Calculate column means
column_means = MD_x.mean()

# Round the column means to two decimal places
rounded_column_means = np.round(column_means, 2)

# Print the rounded column means
print(rounded_column_means)
```

```python
pca = PCA()
MD_pca = pca.fit_transform(MD_x)

# Summarize PCA results
explained_variance_ratio = pca.explained_variance_ratio_
summary = {
    "Standard Deviation": pca.singular_values_,
    "Proportion of Variance": explained_variance_ratio,
    "Cumulative Proportion": explained_variance_ratio.cumsum()
}
print(summary)
```

```python
MD_pca_rounded = np.round(MD_pca, 1)

# Print the rounded PCA results
print(MD_pca_rounded)
```

```python
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], color='grey')
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA Components")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# Set a random seed for reproducibility
np.random.seed(1234)

# Define the range of cluster numbers (2 to 8)
n_clusters_range = range(2, 9)

# Perform KMeans clustering for each number of segments and repetitions
cluster_assignments = {}

for n_clusters in n_clusters_range:
    cluster_results = []

    for _ in range(10):
        kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
        labels = kmeans.fit_predict(MD_x)
        cluster_results.append(labels)

    cluster_assignments[n_clusters] = cluster_results

silhouette_scores = []

for n_clusters, results in cluster_assignments.items():
    avg_silhouette_score = np.mean([silhouette_score(MD_x, labels) for labels in results])
    silhouette_scores.append(avg_silhouette_score)

# Create a plot
plt.plot(n_clusters_range, silhouette_scores, marker='o', linestyle='-')
plt.xlabel("Number of Segments")
plt.ylabel("Average Silhouette Score")
plt.title("Clustering Results")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
np.random.seed(1234)

# Define the range of cluster numbers (2 to 8)
n_clusters_range = range(2, 9)

# Number of bootstrap samples
n_bootstrap_samples = 100

# Initialize lists to store adjusted Rand indices
adjusted_rand_indices = []

for n_clusters in n_clusters_range:
    adjusted_rand_scores = []

    for _ in range(n_bootstrap_samples):
        # Sample with replacement to create a bootstrap sample
        bootstrap_sample_indices = np.random.choice(len(MD_x), len(MD_x), replace=True)
        bootstrap_sample = MD_x.iloc[bootstrap_sample_indices]

        # Perform KMeans clustering on the bootstrap sample
        kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
        labels = kmeans.fit_predict(bootstrap_sample)

        # Compute the adjusted Rand index for the clustering result
        true_labels = np.random.choice(n_clusters, len(bootstrap_sample))
        adjusted_rand = adjusted_rand_score(true_labels, labels)
        adjusted_rand_scores.append(adjusted_rand)

    # Calculate the mean adjusted Rand index for the current number of clusters
    mean_adjusted_rand = np.mean(adjusted_rand_scores)
    adjusted_rand_indices.append(mean_adjusted_rand)

# Create a plot
plt.plot(n_clusters_range, adjusted_rand_indices, marker='o', linestyle='-')
plt.xlabel("Number of Segments")
plt.ylabel("Adjusted Rand Index")
plt.title("Adjusted Rand Index vs. Number of Segments")
plt.grid(True)
plt.show()
```
