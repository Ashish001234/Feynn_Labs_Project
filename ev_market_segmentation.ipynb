```python
numerical_features = ['Battery Capacity(kWh)', 'Acceleration(sec)', 'TopSpeed(km/h)', 'Range(km)', 'Max Power(kW)', 'Max Torque(Nm)', 'No. of Seats', 'Price(Lh)']
sns.pairplot(df[numerical_features])
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
correlation_matrix = df[numerical_features].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand Name', y='Price(Lh)', data=df)
plt.title('Price Distribution by Brand Name')
plt.xticks(rotation=90)
plt.xlabel('Brand Name')
plt.ylabel('Price (Lh)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Drive Type', palette="Set2")
plt.title('Count of Drive Types')
plt.xlabel('Drive Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
